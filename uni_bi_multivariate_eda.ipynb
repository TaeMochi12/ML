## _Univariate, Bivariate and Multivariate Analysis_

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
```

```python
df.head(20)
```

```python
df.shape
```

> ### Univariate Analysis

```python
df_setosa=df.loc[df['species']=='setosa']
df_setosa
```

```python
df_virginica=df.loc[df['species']=='virginica']
df_versicolor=df.loc[df['species']=='versicolor']
```

```python
# Taking just one feature as x axis and taking y all zeros of length same as x axis i.e. df_setossa['sepal_length'] for univariate analysis
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'o')
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']),'o')
plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor['sepal_length']),'o')
plt.xlabel('Sepal Length')
plt.show()
```

```python
# Points are easily indetifiable as follows:
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'o')
```

```python
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'o')
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']),'o')
```

> ### Bivariate Analysis

```python
sns.FacetGrid(df,hue='species',size=5).map(plt.scatter,"petal_length","sepal_width").add_legend()

# It is clear from the plot the setosa species is clearly classified but some points of versicolor and virginica are overlapping
```

> ### Multivariate Analysis

```python
sns.pairplot(df,hue='species',size=2)
```
