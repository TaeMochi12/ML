## MatplotLib Tutorial
Matplotlib is a plotting linrary for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter,wxPython,Qt, or GTK+.<br>
<br>
Pros of Matplotlib:
- Easy for simple plots
- Custom labels and texts
- Great control of every element in the figure
- High-quality output in many formats
- Very customizable in general

```python
import matplotlib.pyplot as plt
#now no need to write plt.show everytime
%matplotlib inline  
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
# Scatter Plot
plt.scatter(x,y,c='b')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
```

```python
y=x*x
```

```python
# Pot
plt.plot(x,y,'ro--')
```

```python
plt.plot(x,y,'b*',linestyle='dashed',linewidth=2,markersize=16)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D Diagram')
```

```python
#Creating Subplots
#Imp parameters: nrows,ncols,position

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*-')
plt.subplot(2,2,3)
plt.plot(x,y,'bo--')

```

```python
x=np.arange(1,11)
y=3*x+5
plt.title("Matplotlib demo")
plt.xlabel("x axis caption")
plt.ylabel("y axis caption")
plt.plot(x,y)
```

```python
#Compute the x and y coordinates for points on a sine curve
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title("sine wave form")

#Plot the points using matplotlib
plt.plot(x,y)
```

```python
# Bar Plot

x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y,align='center')
plt.bar(x2,y2,color='g')
plt.title('Bar Graph')
plt.ylabel("Y axis")
plt.xlabel("X axis")
```
